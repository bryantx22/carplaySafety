```python
import requests
from bs4 import BeautifulSoup

# URL of the website you want to scrape
url = "https://www.apple.com/ios/carplay/available-models/"

# Send a GET request to the URL
response = requests.get(url)
```

```python
brands = []
models = []
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Now you can use BeautifulSoup to extract information from the page
    # For example, let's extract all the <a> tags (links) on the page
    table = soup.find(class_="section section-models")
    rows = table.find_all(class_='row')

    for row in rows:
        ul_element = row.find('ul')
        model = [li.text for li in ul_element.find_all('li')]
        brand = [row.find(class_='image-logo-text').text] * len(model)
        brands.extend(brand)
        models.extend(model)
else:
    print('Failed to retrieve the webpage')
```

```python
start = []
end = []
short = []
models_clean = []
```

```python
for model in models:
    
    model = " ".join(model.split())
    models_clean.append(model)

    string = model.split()
    start.append(float(string[0]))

    if string[1] == "-":
        end.append(float(string[2]))
        short.append(' '.join(string[3:]))
    else:
        end.append(-1)
        short.append(' '.join(string[1:]))
```

```python
import pandas as pd
```

```python
data =  pd.DataFrame({
    'brand': brands,
    'full_name': models_clean, 
    'short': short, 
    'start': start,
    'end': end
 })
```

```python
data.to_csv('apple.csv', index=False)
```
